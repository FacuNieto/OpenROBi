```python
#Carga de dataset y visualizaci√≥n
df2=pd.read_csv("OpenBCI-RAW-2022-04-24_21-26-52.txt",sep=', ', engine='python')
# print()
index= np.arange(0,100000,1) #index=el rango que queremos visualizar
df2= pd.DataFrame(df2, columns= ['c1','c2','c3','c4','c5','c6','c7'],index=index)
plt.rcParams["figure.dpi"] = 300 #calidad de imagen
plt.close('all')
df2.plot(figsize=(12,5.8),linewidth=0.5,subplots=True) #linewidth es el ancho de la linea
plt.grid() #grilla
plt.legend(loc=1) #posici√≥n de los nombres
plt.savefig('asdf.png') #guardamos
plt.show()
```

```python
df2n=np.array(df2)
df2n =df2n.transpose()
df2n[0]
for i in range(0,7):
    DataFilter.detrend(df2n[i], DetrendOperations.CONSTANT.value)
    DataFilter.perform_bandpass(df2n[i], 250,
                                5, 40, 4,
                                FilterTypes.BUTTERWORTH.value, 0)
    DataFilter.perform_bandstop(df2n[i], 250,
                                50.0, 4.0, 2,
                                FilterTypes.BUTTERWORTH.value, 0)
    DataFilter.perform_bandstop(df2n[i], 250,
                                100.0, 4.0, 2,
                                FilterTypes.BUTTERWORTH.value, 0)
    # DataFilter.perform_bandpass(data[1], 250, 5, 45, 3,
                                            # FilterTypes.BUTTERWORTH.value, 0)


df2n =df2n.transpose()
index= np.arange(0,100000,1) #index=el rango que queremos visualizar
df2n= pd.DataFrame(df2n, columns= ['c1','c2','c3','c4','c5','c6','c7'],index=index)
plt.rcParams["figure.dpi"] = 300 #calidad de imagen
plt.close('all')
df2n.plot(figsize=(12,5.8),linewidth=0.5,subplots=1) #linewidth es el ancho de la linea
plt.grid() #grilla
plt.legend(loc=1) #posici√≥n de los nombres
plt.savefig('asdf1.png') #guardamos
plt.show()

# qubernetes, base de datos, balance de cargas
```
